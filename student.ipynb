```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

```

```python
df = pd.read_csv('Student_Performance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize= (4,5))
ax = sns.countplot(data=df, x = df['gender'])
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sns.scatterplot(x='study_hours', y='overall_score', data=df)
plt.title("Study Hours vs Overall Score")
plt.show()
```

```python
sns.scatterplot(x='attendance_percentage', y='overall_score', data=df)
plt.title("Attendance vs Score")
plt.show()
```

```python
sns.barplot(x='gender', y='overall_score', data=df)
plt.title("Average Score by Gender")
plt.show()
```

```python
sns.countplot(x='final_grade', data=df)
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure()
df[['math_score', 'science_score', 'english_score']].mean().plot(kind='bar')
plt.title("Average Score by Subject")
plt.show()
```

```python
plt.figure()
sns.barplot(x='school_type', y='overall_score', data=df)
plt.title("School Type vs Performance")
plt.show()
```

```python
plt.figure()
sns.barplot(x='internet_access', y='overall_score', data=df)
plt.title("Internet Access vs Score")
plt.show()
```

```python
plt.figure()
sns.barplot(x='parent_education', y='overall_score', data=df)
plt.xticks(rotation=45)
plt.title("Parent Education vs Score")
plt.show()
```

```python
plt.figure()
sns.barplot(x='travel_time', y='overall_score', data=df)
plt.title("Travel time vs Score")
plt.show()
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score']

df['result'] = df['overall_score'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
```

```python
top_students = df[df['overall_score'] > 80]
low_students = df[df['overall_score'] < 40]

print("\nTop Students Avg Study Hours:", top_students['study_hours'].mean())
print("Low Students Avg Study Hours:", low_students['study_hours'].mean())
```

```python
gb = df.groupby("parent_education").agg({"math_score":'mean', "science_score":'mean', "english_score":'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot=True)
plt.show()
```

```python
gb = df.groupby("attendance_percentage").agg({"math_score":'mean', "science_score":'mean', "english_score":'mean'})
print(gb)
```

```python
gb = df.groupby("extra_activities").agg({"math_score":'mean', "science_score":'mean', "english_score":'mean'})
print(gb)
```
